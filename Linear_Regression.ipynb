```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("homeprices.csv")
df
```

```python
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area, df.price, color='red', marker='+')
```

```python
df[['area']]
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df.price)
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
d = pd.read_csv("areas.csv")
d
```

```python
price = reg.predict(d)
```

```python
d['prices'] = price
```

```python
d
```

```python
d.to_csv("prediction.csv", index=False)
```

```python
plt.xlabel('area', fontsize=20)
plt.ylabel('price', fontsize=20)
plt.scatter(df.area, df.price, color='red', marker='+')
plt.plot(df.area, reg.predict(df[['area']]), color='blue')
```

Working on Canada per capita Income


```python
cpi = pd.read_csv("canada_per_capita_income.csv")
cpi.head()
```

```python
plt.xlabel('year', fontsize=20)
plt.ylabel('per capita income (US$)', fontsize=20)
plt.scatter(cpi.year, cpi.per_capita_income_USD, color='red', marker='+')
```

```python
# model with linear regression
reg = linear_model.LinearRegression()
reg.fit(cpi[['year']], cpi.per_capita_income_USD)
```

```python
reg.predict([[2020]])
```

```python
# make a slope on graph with linear regression
plt.xlabel('year', fontsize=20)
plt.ylabel('per capita income (US$)', fontsize=20)
plt.scatter(cpi.year, cpi.per_capita_income_USD, color='red', marker='+')
plt.plot(cpi.year, reg.predict(cpi[['year']]), color='blue')    
```
